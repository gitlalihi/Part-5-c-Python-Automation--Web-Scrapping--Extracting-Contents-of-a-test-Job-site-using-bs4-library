```python
import requests
from bs4 import BeautifulSoup
#import time- to handle time between two requests as a method for best practice

url="https://realpython.github.io/fake-jobs/"
#delay_seconds=5

response=requests.get(url)
#if response.status_code==200:

html_content=response.text
soup=BeautifulSoup(html_content,'html.parser')

# to find the object holding job listings as per retrived html , we find the container- div id= Result Container
    
job_results=soup.find(id='ResultsContainer')
print(job_results.prettify())

#time.sleep(delay_seconds)
#else:
#print(f"Response error occured:{response.status_code}")
        
    
```

```python
# to find all job posts, each of all jobs are under- 'div class = card-content'# use find_all
job_ele=soup.find_all('div',class_="card-content")
for i in job_ele:
    print(i,end="\n")
```

```python
# Extract the elements of job position,Company name ,location and date of posting
job_ele=soup.find_all('div',class_="card-content")

for y in job_ele:
    title_element = y.find("h2", class_="title is-5")
    company_element = y.find("h3", class_="subtitle is-6 company")
    location_element = y.find("p", class_="location")
    date_of_post=y.find("p",class_="is-small has-text-grey")
    
    print( "Job Title: ",title_element.text.strip())
    print("Comapny name: ",company_element.text.strip())
    print("Location: " ,location_element.text.strip())
    print("Date of posting:",date_of_post.text.strip())
    print()
```

```python
# to extract  only" Product manager" jobs
manager_jobs=soup.find_all("h2",string="Product manager")
print(manager_jobs)
```

```python
# to find acess the parents and grandparents
python_jobs = soup.find_all(
    "h2", string=lambda text:"python"in text.lower() )

python_job_elements = [
    h2_element.parent.parent.parent for h2_element in python_jobs
]
print(python_job_elements)
```

```python
# to acess the jobs from the parent elements and get specific jobs
python_jobs = soup.find_all(
    "h2", string=lambda text:"officer"in text.lower() )

python_job_parent_elements = [
    h2_element.parent.parent.parent for h2_element in python_jobs
]
for job_elements in python_job_parent_elements:
    title_element = job_elements.find("h2", class_="title is-5")
    company_element = job_elements.find("h3", class_="subtitle is-6 company")
    location_element = job_elements.find("p", class_="location")
    date_of_post=job_elements.find("p",class_="is-small has-text-grey")
    
    print( "Job Title: ",title_element.text.strip())
    print("Comapny name: ",company_element.text.strip())
    print("Location: " ,location_element.text.strip())
    print("Date of posting:",date_of_post.text.strip())
    print()
```

```python
# to acess via children by going down the tree
python_jobs = soup.find_all(
    "h2", string=lambda text: "manager" in text.lower()
)

for job_element in python_jobs:
    title_element = job_element.find_next("h2", class_="title is-5")
    company_element = job_element.find_next("h3", class_="subtitle is-6 company")
    location_element = job_element.find_next("p", class_="location")
    date_of_post = job_element.find_next("p", class_="is-small has-text-grey")

    if title_element and company_element and location_element and date_of_post:
        print("Job Title: ", title_element.text.strip())
        print("Company name: ", company_element.text.strip())
        print("Location: ", location_element.text.strip())
        print("Date of posting:", date_of_post.text.strip())
        print()
```

```python
# To extract links for application of jobs from the html
job_ele=soup.find_all('footer',class_="card-footer")
for job_element in job_ele:
    link_url = job_element.find_all("a")[1]["href"]
    print(f"Apply here: {link_url}\n")
    
```
